```python
#%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "../data/city_data.csv"
ride_data_to_load = "../data/ride_data.csv"

# Read the City and Ride Data
city_df = pd.read_csv(city_data_to_load)
ride_df = pd.read_csv(ride_data_to_load)
#city_df.head()
#ride_df.head()
# Combine the data into a single dataset
cityride_df = city_df.merge(ride_df, on='city', sort=True)

# Display the data table for preview
cityride_df.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
list_citytypes = cityride_df['type'].unique()
colors = ['gold', 'lightblue', 'coral']
for citytype in list_citytypes:
    type_df = cityride_df[cityride_df['type'] == citytype]
    list_cities = type_df['city'].unique()
    i = 0
    list_columns = ["city", "avg_fare", "no_rides", "no_drivers"]
    scatter_df = pd.DataFrame(columns=list_columns)
    for city in list_cities:
        city_df = type_df[type_df['city'] == city]
        avg_fare = city_df['fare'].mean()
        no_rides = city_df['ride_id'].nunique()
        no_drivers_list = city_df['driver_count'].unique()
        no_drivers = no_drivers_list[0]
        list_results = [city, avg_fare, no_rides, no_drivers]
        scatter_df.loc[i] = list_results
        i += 1
    #print(scatter_df.head())
    #print(scatter_df['no_drivers'])
    #print(len(scatter_df['no_drivers']))
    #print(scatter_df.iloc[0][3])
    s = [scatter_df.iloc[n][3]*10 for n in range(len(scatter_df['no_drivers']))]
    #print(s)
    plt.scatter(scatter_df['no_rides'],scatter_df['avg_fare'], s=s, facecolors=colors, edgecolors="black", edgecolor='w', alpha=0.6,cmap=colors)
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")
plt.title("Pyber Ride Sharing Data (2016)")
plt.legend(list_citytypes,title="City Types",markerscale=1)
plt.text(41,35,"Note:")
plt.text(41,33,"Circle size correlates with driver count per city")
plt.show()
#print(scatter_df['no_drivers'])
#print(i)
#scatter_df.head(18)

# Build the scatter plots for each city types

# Incorporate the other graph properties

# Create a legend

# Incorporate a text label regarding circle size

# Save Figure
```
